**Problem Statement:-**

Write a Python Program for magic square. A magic square is an n * n matrix of the integers 1 to n2 such that the sum of each row, column, and diagonal is the same. The figure given below is an example of magic square for case n=5. In this example, the common sum is 65. 

Capture.PNG



---



```python
import numpy as np
n = int(input("Enter n dimention of the matrix:-"))
matrix = np.zeros((n,n) , dtype=int)

i , j = n//2 , n-1
matrix[i][j] =  1
for k in range(2 , (n*n)+1):  
  i , j = (i-1) , (j+1)

  if (i == -1) and (j == n):
    i , j = 0 , n-2

  else:
    if i == -1:
      i = n-1

    if j == n:
      j = 0
    
  if matrix[i][j] != 0:
    i , j = i+1 , j-2
  
  matrix[i][j] =  k
print(f"A magic square of size an {n}x{n}:-")
print(matrix)
```
